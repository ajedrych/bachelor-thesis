### Analiza mobilności dla Warszawy według danych Apple

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
mobility = pd.read_csv('applemobilitytrends-2022-03-17.csv')
```

```python
mobility
```

```python
mobility.drop(mobility.loc[mobility['region']!='Warsaw'].index, inplace=True)
mobility
```

```python
mobility.to_excel("warszawa_apple.xlsx")
```
